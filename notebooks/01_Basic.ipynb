## Univariate Analysis

#### Central Tendency

```python
# Create new column 'percent_hike'
lr_df['percent_hike'] = (lr_df['salary_2018'] - lr_df['salary_2017']) / lr_df['salary_2017'] * 100
```

```python
lr_df.head()
```

```python
# mean
avg_percent_hike = lr_df['percent_hike'].mean()
print("Mean of Percent Hike:", avg_percent_hike)
```

```python
# median
med_percent_hike = lr_df['percent_hike'].median()
print("Median of Percent Hike:", med_percent_hike)
```

```python
# Value counts (mode)
lr_df['percent_hike'].value_counts()
```

```python
lr_df['percent_hike'].mode()
```

#### Measures of variability

```python
# Range
range_salary_2018 = lr_df['salary_2018'].max() - lr_df['salary_2018'].min()
print("Range of Salary in 2018 is", range_salary_2018)
```

```python
# Interquartile Range
# IQR = Q3-Q1
interquartile_range = lr_df['salary_2018'].quantile(0.75) - lr_df['salary_2018'].quantile(0.25)
interquartile_range
```

```python
# Variance
lr_df['salary_2018'].var()
```
